```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
browser = webdriver.Chrome()
```

```python
browser.get("https://www.youtube.com/@GeeksforGeeksVideos/videos")
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
# sp = soup.find_all('ytd-rich-grid-renderer')
```

```python
data = []

for sp in soup.find_all('ytd-rich-item-renderer'):
    video_link = sp.find('a').get('href')
    # video_link = 'https://www.youtube.com' + video_link
    print(video_link)

    video_title = sp.find('h3').text
    print(video_title)

    try:
        img_link = sp.find('img').get('src').split('?')[0]
        print(img_link)
    except:
        img_link = 'Img not Available'

    video_time_frame = sp.find_all('span')[2].text.replace('\n', '').strip()
    print(video_time_frame)
    
    views = sp.find('span', class_ = "inline-metadata-item style-scope ytd-video-meta-block").text
    print(views)

    days = sp.find('div',class_ = "style-scope ytd-video-meta-block").find_all('span')[2].text
    print(days)
        
    print("----------------------")
    print(' ')
    
    data.append([video_title, views,days, video_time_frame, img_link, video_link])
    

pd.DataFrame(data, columns=['video_title','views','days','video_time_frame','img_link','video_link']).to_csv('youtubes.csv', index=False)
```
