```python
import requests
from bs4 import BeautifulSoup
import time
from tqdm import tqdm
from selenium import webdriver
import pandas as pd
```

```python
browser = webdriver.Chrome()
```

```python
browser.get('https://stock-pictures.netlify.app/')
```

```python
#Infinte Scroll
for i in tqdm(range(0,10000000,1000)):
    browser.execute_script("window.scrollTo(0,"+ str(i) +")")
    time.sleep(.1)
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')   
soup.text
```

```python
data = []
for sp in soup.find_all('div', class_ = "container"):
    img_src = sp.find('img').get('src')
    tags = sp.find('div', class_ = "tags").text[7:]
    likes = sp.find('div', class_ = 'likes-comments').find_all('span')[0].text[:-6]
    comments = sp.find('div', class_ = 'likes-comments').find_all('span')[1].text[:-9]
    data.append([img_src,tags,likes,comments])
```

```python
df = pd.DataFrame(data, columns=['img_link','tags','likes','comments'])
df.head()
```

```python
print(len(soup.find_all('div', class_ = "container")))
```

```python
df = df.drop_duplicates()
```
